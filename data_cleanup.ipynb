```python
#Import neccessary libraries
import pandas as pd
import json
```

**Match Results**

```python
#opening the json file
with open ('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

#converting to a DataFrame
df_match_results = pd.DataFrame(data[0]['matchSummary'])
df_match_results.head() 
```

```python
#No of rows and columns
df_match_results.shape
```

```python
''' In the DataFrame , Scorecard can be identyfied as the unique column. So it can be considered as the 
    Primarykey
'''
#Converting the column scorecard into matchid

df_match_results.rename({'scorecard':'match_id'}, axis=1 , inplace = True)
df_match_results.head()
```

```python
# Exporting the MatchResult DataFrame into csv
df_match_results.to_csv('t20_csv_files/processed_match_results.csv', index=False)
```

```python
# Creating a Key:Value pair using, Team1 and Team2 as KEY and match_id as VALUE.

# Note: Here we create the reverse version of Team2 Vs Team1. 

match_id_dict = {}

for index , row in df_match_results.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']


match_id_dict
```

**Batting Summary**

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)

    # data list has a sublist of battingSummary which groups up all the match between specific two teams
    
    # Looping through the battingSummary records and adding all the records into one list
    all_records = []
    for records in data:
        all_records.extend(records['battingSummary'])

#Coverting the merged records into dataframe
df_batting_summary = pd.DataFrame(all_records)
df_batting_summary.head()
    
```

```python
#Creating a new column 'Out/Not_Out' based on the disimissal column

df_batting_summary['out/not_out'] = df_batting_summary.dismissal.apply(lambda x : 'out' if len(x) > 0 else 'not_out')
df_batting_summary
```

```python
# Removing the dismissal column from the dataframe
df_batting_summary.drop(columns=["dismissal"], inplace=True)
```

```python
df_batting_summary.head(11)

# List of all the batsmanName
# batsman = df_batting_summary['batsmanName'].unique().tolist()


#Removing special characters from the BatsmanName coloumn
df_batting_summary['batsmanName'] = df_batting_summary['batsmanName'].apply(lambda x : x.replace('â€' , ''))
df_batting_summary['batsmanName'] = df_batting_summary['batsmanName'].apply(lambda x : x.replace('\xa0' , ''))

df_batting_summary
```

```python
'''The Common ground between "Match Result" dataframe and "BattingSummary" dataframe can be build using the
dictionary match_id_dict. '''

df_batting_summary['match_id'] = df_batting_summary['match'].map(match_id_dict)
df_batting_summary
```

```python
# Exporting the DataFrame as a CSV file

df_batting_summary.to_csv('t20_csv_files/processed_batting_summary.csv', index=False)
```

Processing Bowling Sumamry

```python
df_bowling_summary.to_csv('t20_csv_files/processed_bowling_summary.csv', index = False)
```
